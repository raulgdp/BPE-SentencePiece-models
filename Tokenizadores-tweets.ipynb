# Regex con nltk tokinazaciÃ³n para tweets: TweetTokenizer y regexp_tokenize  se importan desde nltk.tokenize

```python
tweet2=['Tw33t # @dude_really in (01/02/06)  #hash_tag $hit (g@y) retard @dude. ðŸ˜€ðŸ˜€ !ðŸ˜€abc %ðŸ˜€lol #hateit #hate.it $%&/ f*ck- in love_twitter']
```

```python
from nltk.tokenize import regexp_tokenize, TweetTokenizer

# Define a regex pattern to find hashtags: pattern1
patron1 = r"#\w+"

# Use the pattern on the first tweet in the tweets list
hashtags = regexp_tokenize(tweet2[0], patron1)
print(hashtags)
```

```python
pattern2 = r"[@|#]\w+"
mentions_hashtags = regexp_tokenize(tweet2[-1], pattern2)
print(mentions_hashtags)
```

```python
tknzr = TweetTokenizer()
all_tokens = [tknzr.tokenize(t) for t in tweet2]
print(all_tokens)
```

# Reconocemos el hastag adicionando una regla con matcher

```python
import spacy
from spacy.matcher import Matcher  #librerÃ­a para expresiones regulares
from spacy.tokens import Token

nlp = spacy.load("es_core_news_sm")
matcher = Matcher(nlp.vocab)

# Add pattern for valid hashtag, i.e. '#' plus any ASCII token
matcher.add("HASHTAG", [[{"ORTH": "#"}, {"IS_ASCII": True}]])

# Register token extension
Token.set_extension("is_hashtag", default=False)

doc = nlp("Hello world ðŸ˜€ #MondayMotivation")
#doc = nlp("La ministra @VictoriaBxE anuncia que el 100% de los colegios y niÃ±os deberÃ¡n estar abiertos y en alternancia entre abril y mayo. Excelente noticia! #LaEducacionPresencialEsVital @constanzaalarc @dario_maldonado @sandragarciajar @MSantamariaS @patolinecamacho @julrubiano")
matches = matcher(doc)
hashtags = []
for match_id, start, end in matches:
    if doc.vocab.strings[match_id] == "HASHTAG":
        hashtags.append(doc[start:end])
with doc.retokenize() as retokenizer:
    for span in hashtags:
        retokenizer.merge(span)
        for token in span:
            token._.is_hashtag = True

for token in doc:
    print(token.text, token._.is_hashtag)
```
